## Questions

__1. How old were the youngest male and female participants of the 1996 Olympics?__

- 16 and 15
- 14 and 12 
- 16 and 12
- 13 and 11

```python
male = data[(data['Sex'] == 'M') & (data['Year'] == 1996)]['Age'].min()
female = data[(data['Sex'] == 'F') & (data['Year'] == 1996)]['Age'].min()
print(male, 'and', female)
```

__2. What was the percentage of male gymnasts among all the male participants of the 2000 Olympics? Round the answer to the first decimal.__

*Hint:* here and further if needed drop duplicated sportsmen to count only unique ones. 

- 0.2
- 1.5 
- 2.5
- 7.7

```python
male_gym = data[(data['Sex'] == 'M') & (data['Year'] == 2000) & (data['Sport'] == 'Gymnastics')]['ID'].nunique()
male_part = data[(data['Sex'] == 'M') & (data['Year'] == 2000)]['ID'].nunique()
round(male_gym / male_part * 100, 1)
```

Answer:

***B) 1.5***

__3. What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? Round the answer to the first decimal.__

- 178.5 and 7.2
- 179.4 and 10
- 180.7 and 6.7
- 182.4 and 9.1 

```python
height = data[(data['Sex'] == 'F') & (data['Sport'] == 'Basketball') & (data['Year'] == 2000)]['Height']
print(round(height.mean(), 1), 'and', round(height.std(), 1))
```

Answer:

***D) 182.4 and 9.1***

__4. Find a sportsperson participated in the 2002 Olympics, with the highest weight among other participants of the same Olympics. What sport did he or she do?__

- Judo
- Bobsleigh 
- Weightlifting
- Boxing

```python
data[data.Year == 2002].groupby(by=['Year','Season'],as_index=False)['ID'].count()
max_Weight = data[data.Year == 2002]['Weight'].max()
data.loc[(data.Year == 2002)&(data.Weight == max_Weight),:]
data.loc[(data.Year == 2002)&(data.Weight == max_Weight),:].Sport
```

Answer:

** B) Bobsleigh**

__7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 

- Yes 
- No

```python
def n_medal(country): 
    print(country,':',data[(data.Year==2016)&(data.Team==country)&(data.Medal.notnull())]['Medal'].count(),"medals")
```

```python
n_medal('Switzerland')
n_medal('Serbia')
```

Answer:

**A) Yes**

__8. What age category did the fewest and the most participants of the 2014 Olympics belong to?__
- [45-55] and [25-35) correspondingly
- [45-55] and [15-25) correspondingly
- [35-45] and [25-35) correspondingly
- [45-55] and [35-45) correspondingly

```python
age_15 = data[(data['Year'] == 2014) & (data['Age'] >= 15) & (data['Age'] < 25)]['ID'].nunique()
age_25 = data[(data['Year'] == 2014) & (data['Age'] >= 25) & (data['Age'] < 35)]['ID'].nunique()
age_35 = data[(data['Year'] == 2014) & (data['Age'] >= 35) & (data['Age'] <= 45)]['ID'].nunique()
age_45 = data[(data['Year'] == 2014) & (data['Age'] >= 45) & (data['Age'] <= 55)]['ID'].nunique()
ages = {age_35:'35-45', age_45:'45-55'}
print(ages[min(age_35, age_45)], end=' ')
age_35 = data[(data['Year'] == 2014) & (data['Age'] >= 35) & (data['Age'] < 45)]['ID'].nunique()
ages = {age_15:'15-25', age_25:'25-35', age_35:'35-45'}
print('and', ages[max(age_15, age_25, age_35)])
```

Answer:

**A) [45, 55] and [25 , 35) **
